```python
from collections import defaultdict

from Pipeline.Global.Plotting import Plotting
from Pipeline.Methodology import ABC_Testing
```

```python
all_trace_record, result_summary = ABC_Testing.lcb_trace_evaluation("")
```

```python
mi_trace_list = [(10, 120), (20, 120), (50, 120), (80, 120)]
sn_trace_list = [(80, 100), (80, 120), (80, 150), (80, 250)]
cross_trace_list = [(10, 50), (100, 250), (10, 1000), (20, 1000), (100, 120), (80, 120)]

df_dict = defaultdict(list)

mi_trace_dataframe = []
sn_trace_dataframe = []
cross_trace_dataframe = []
```

```python
for df in all_trace_record:
    if df.empty:
        continue

    sn_val = int(df['Solution_Size'].iloc[0])
    mi_val = int(df['Max_Iteration'].iloc[0])
    current_config = (sn_val, mi_val)

    df_dict[current_config].append(df)

mi_trace_dataframe = [df for config in mi_trace_list for df in df_dict[config]]
sn_trace_dataframe = [df for config in sn_trace_list for df in df_dict[config]]
cross_trace_dataframe = [df for config in cross_trace_list for df in df_dict[config]]
```

```python
Plotting.plot_2x2_experiment_grid(mi_trace_dataframe, (0.475, 0.7), (0, 4.5))
```

```python
Plotting.plot_2x2_experiment_grid(sn_trace_dataframe, (0.475, 0.7), (0, 4.5))
```

```python
Plotting.plot_3x2_experiment_grid(cross_trace_dataframe, (0.475, 0.7), (0, 4.5))
```
