```python
import pandas as pd
```

```python
def extract_best_seed(df):
    numeric_cols = df.select_dtypes(include=['number']).columns
    stats = df.groupby('Seed')[numeric_cols].agg(['mean', 'std'])

    stats.columns = [f"{col}_{stat}" for col, stat in stats.columns]
    stats['MCC_LCB'] = stats['MCC_mean'] - stats['MCC_std']

    target_columns = [
        'Accuracy_mean', 'Accuracy_std',
        'F1-Score_mean', 'F1-Score_std',
        'MCC_mean', 'MCC_std', 'MCC_LCB'
    ]
    best_result = stats.reset_index()[['Seed'] + target_columns]
    best_result = best_result.sort_values(['MCC_LCB'], ascending=False).iloc[[0]]
    return best_result
```

```python
abc_relm_cv = pd.read_csv('cleaned_ABC_RELM_CV_Algo_3_2_80SN_120MI_Results.csv')
extract_best_seed(abc_relm_cv)
```

```python
relm = pd.read_csv('cleaned_Grid_Optimization_Results.csv')
extract_best_seed(relm)
```
