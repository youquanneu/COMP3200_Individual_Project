```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler

from Pipeline.Algorithm.ExtremeLearningMachine import ExtremeLearningMachine
from Pipeline.Global.GlobalSetting import GlobalSetting
from Pipeline.Methodology.EvaluationMatrix import EvaluationMatrix
```

```python
df_raw = pd.read_csv('../../Storage/Dataset/UCI_Gallstone_Dataset.csv')
df_clear = pd.read_csv('../../Storage/Dataset/Full_Clean_Dataset.csv')
target_col = ['Gallstone Status']
```

```python
x_raw = df_raw.drop(target_col, axis=1)
feature_size = x_raw.shape[1]
y_raw = df_raw[target_col]
X_train, X_test, y_train, y_test = train_test_split(
    x_raw,
    y_raw,
    test_size=0.2,
    random_state=42
)
scaler = MinMaxScaler()
X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
X_train_scaled_df = pd.DataFrame(X_train_scaled, columns=x_raw.columns)
X_test_scaled_df = pd.DataFrame(X_test_scaled, columns=x_raw.columns)
```

```python
x_clear = df_clear.drop(target_col, axis=1)
y_clear = df_clear[target_col]
X_train_clear, X_test_clear, y_train_clear, y_test_clear = train_test_split(
    x_clear,
    y_clear,
    test_size=0.2,
    random_state=42
)
scaler_clear = MinMaxScaler()
X_train_clear_scaled = scaler_clear.fit_transform(X_train_clear)
X_test_clear_scaled = scaler_clear.transform(X_test_clear)
X_train_clear_scaled = pd.DataFrame(X_train_clear_scaled, columns=x_clear.columns)
X_test_clear_scaled = pd.DataFrame(X_test_clear_scaled, columns=x_clear.columns)
```

```python
elm_raw = ExtremeLearningMachine(feature_size,32,GlobalSetting.sigmoid,random_seed=42)
elm_raw.fit(X_train_scaled_df,y_train)
y_pred_raw = elm_raw.predict(X_test_scaled_df)
eval_raw = EvaluationMatrix(y_true= y_test, y_pred= y_pred_raw)
eval_raw.get_report()
```

```python
elm_clear = ExtremeLearningMachine(feature_size,32,GlobalSetting.sigmoid,random_seed=42)
elm_clear.fit(X_train_clear_scaled,y_train_clear)
y_pred_clear = elm_clear.predict(X_test_clear_scaled)
eval_clear = EvaluationMatrix(y_true= y_test_clear, y_pred= y_pred_clear)
eval_clear.get_report()
```
