```python
import pandas as pd

from Pipeline.Global.PlottingData import PlottingData
```

```python
filePath = '../../Storage/Dataset/UCI_Gallstone_Dataset.csv'
df = pd.read_csv(filePath)
```

```python
df_numeric = df.select_dtypes(include=['number', 'bool']).drop(columns=['Gallstone Status'], errors='ignore')
target_corr = df_numeric.corrwith(df['Gallstone Status']).dropna().sort_values(ascending=True)
corr_df = target_corr.reset_index()
corr_df.columns = ['Feature', 'Pearson_Correlation']
corr_df['Pearson_Correlation'] = corr_df['Pearson_Correlation'].round(4)
corr_df
```
