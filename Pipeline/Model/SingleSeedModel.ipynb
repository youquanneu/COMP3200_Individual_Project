```python
import pandas as pd
from sklearn.preprocessing import MinMaxScaler

from Pipeline.Algorithm.ArtificialBeeColonyElm import ArtificialBeeColonyElm
from Pipeline.Algorithm.ArtificialBeeColonyElmCVEnsemble import ArtificialBeeColonyElmCVEnsemble
from Pipeline.Algorithm.ExtremeLearningMachine import ExtremeLearningMachine
from Pipeline.Global.GallstoneDataSet import GallstoneDataSet
from Pipeline.Global.GlobalSetting import GlobalSetting
from Pipeline.Methodology.EvaluationMatrix import EvaluationMatrix
```

```python
random_seed = 42
cv_test_count = 5
```

```python
gallstone_dataset = GallstoneDataSet()
gallstone_dataset.fetch_data_path_0()
gallstone_dataset.cross_validate_test(cv_test_count)
```

```python
model_configs = GlobalSetting.get_model_configs()
```

```python
elm_config = next(cfg for cfg in model_configs if cfg["Model_Types"] == "Best_Hidden_Nodes")
elm_fold_results = []
for fold_idx in range(cv_test_count):
    x_train, y_train, x_test, y_test = gallstone_dataset.fold_split[fold_idx]
    features_size = x_train.shape[1]
    scaler  = MinMaxScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled  = scaler.transform(x_test)
    elm = ExtremeLearningMachine(features_size=features_size,
                                 hidden_size=elm_config['Hidden_Nodes'],
                                 activation_function=elm_config['Activation'],
                                 regularization_lambda=elm_config['Lambda_Value'])
    elm.initialize_random_weights(scale=1.0, random_seed = 42)
    elm.fit(x_train_scaled,y_train)
    y_pred = elm.predict(x_test_scaled)
    evaluation = EvaluationMatrix(y_true=y_test,y_pred=y_pred)
    elm_fold_results.append(evaluation.get_all_metrics())
elm_final_result = pd.DataFrame(elm_fold_results)
elm_final_result.describe()
```

```python
relm_config = next(cfg for cfg in model_configs if cfg["Model_Types"] == "Best_Lambda")
relm_fold_results = []
for fold_idx in range(cv_test_count):
    x_train, y_train, x_test, y_test = gallstone_dataset.fold_split[fold_idx]
    features_size = x_train.shape[1]
    scaler  = MinMaxScaler()
    x_train_scaled = scaler.fit_transform(x_train)
    x_test_scaled  = scaler.transform(x_test)
    relm = ExtremeLearningMachine(features_size         = features_size,
                                  hidden_size           = relm_config['Hidden_Nodes'],
                                  activation_function   = relm_config['Activation'],
                                  regularization_lambda = relm_config['Lambda_Value'])
    relm.initialize_random_weights(scale=1.0, random_seed = 42)
    relm.fit(x_train_scaled,y_train)
    y_pred = relm.predict(x_test_scaled)
    evaluation_relm = EvaluationMatrix(y_true=y_test,y_pred=y_pred)
    relm_fold_results.append(evaluation_relm.get_all_metrics())
relm_final_result = pd.DataFrame(relm_fold_results)
relm_final_result.describe()
```
